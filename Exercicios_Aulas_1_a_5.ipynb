* `receita_anual = 250000`

* `despesa_variavel_mensal = 5000`
* `depreciacao_anual = 12000`

Sabendo disso:

1. Calcule a despesa operacional anual somando `custo_operacional_mensal * 12` e `despesa_variavel_mensal * 12`.
2. Subtraia da `receita_anual` a despesa operacional anual e a `depreciacao_anual`.

Um analista precisa calcular o indicador EBITDA (Earnings Before Interest, Taxes, Depreciation and Amortization) para um conjunto de dados simplificado. As informações são:

* `receita_trimestral = 800000`
* `custo_produtos = 300000`
* `despesa_operacional = 100000`
* `despesa_financeira = 20000`
* `impostos = 50000`
* `depreciacao = 30000`
* `amortizacao = 10000`

Especifique as variáveis correspondentes, e depois:

1. Calcule o lucro operacional bruto (`receita_trimestral - custo_produtos - despesa_operacional`).
2. Subtraia `despesa_financeira` e `impostos` desse lucro operacional bruto para obter o lucro líquido antes de depreciação e amortização.
3. Por fim, calcule o EBITDA somando `depreciacao` e `amortizacao` ao valor obtido no passo anterior. Armazene em `ebitda_trimestral`.

---


````python
receita_trimestral = 800000
custo_produtos = 300000
despesa_operacional = 100000
despesa_financeira = 20000
impostos = 50000
depreciacao = 30000
amortizacao = 10000

lucro_operacional_bruto = receita_trimestral - custo_produtos - despesa_operacional
lucro_liquido_antes_depreciacao = lucro_operacional_bruto - despesa_financeira - impostos
ebitda_trimestral = lucro_liquido_antes_depreciacao + depreciacao + amortizacao
print(ebitda_trimestral)
````

## Diferença entre `sort()` e `sorted()`

### `lista.sort()`
- Método de listas
- Modifica a lista original
- Não retorna nova lista (`return None`)

```python
nums = [3, 1, 2]

nums.sort()

print(nums)
# [1, 2, 3]

```

### `sorted()`
- Função do Python
- Não altera o original
- Retorna uma nova lista ordenada
- Funciona com qualquer iterável (lista, tupla, string, etc.)
```python

nums = [3, 1, 2]

ordenados = sorted(nums)

print(nums)
# [3, 1, 2]

print(ordenados)
# [1, 2, 3]
```





















**Exercício 5:**
Você recebeu um dicionário chamado `dados_vendas` que contém, para cada mês do ano, o valor total de vendas. Por exemplo:

```python
dados_vendas = {
    "janeiro": 12000.0,
    "fevereiro": 15000.0,
    "março": 13000.0,
    "abril": 14000.0,
    "maio": 16000.0
}
```

1. Adicione ao dicionário o mês de `junho` com valor de 17000.0.
2. Calcule o total anual (soma de todos os valores presentes no dicionário).
3. Identifique mês com maior venda e mês com menor venda, armazenando os nomes desses meses em `mes_maior_venda` e `mes_menor_venda`. Dica: Você pode utilizar a função `max()` para encontrar a chave que possui o maior valor; passe o parâmetro `key=dados_vendas.get` para que o critério de comparação seja o valor correspondente a cada chave. De forma semelhante, use `min()` com `key=dados_vendas.get` para obter a chave de menor valor.
4. Crie uma lista chamada `meses` que contenha apenas as chaves (meses) do dicionário, e uma lista chamada `valores` que contenha apenas os valores de vendas.
5. Exiba todos esses resultados (total anual, `mes_maior_venda`, `mes_menor_venda`, `meses`, `valores`) usando prints adequados.



````python
dados_vendas = {
    "janeiro": 12000.0,
    "fevereiro": 15000.0,
    "março": 13000.0,
    "abril": 14000.0,
    "maio": 16000.0
}
dados_vendas["junho"] = 17000.0
total_anual = sum(dados_vendas.values())
mes_maior_venda = max(dados_vendas, key=dados_vendas.get)
mes_menor_venda = min(dados_vendas, key=dados_vendas.get)
meses = list(dados_vendas.keys())
valores = list(dados_vendas.values())
print(f"Total anual: {total_anual}")
print(f"Mês com maior venda: {mes_maior_venda}")
print(f"Mês com menor venda: {mes_menor_venda}")
print(f"Meses: {meses}")
print(f"Valores: {valores}")
````
